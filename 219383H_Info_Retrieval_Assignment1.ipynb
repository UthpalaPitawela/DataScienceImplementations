Read text from student feedback, research paper and twitter texts

```python
def read_file(filename):
    "Read the contents of FILENAME and return as a string."
    infile = open(filename) # windows users should use codecs.open after importing codecs
    contents = infile.read()
    infile.close()
    return contents
```

```python
student_feedback =  read_file("/content/student_feedback.txt")
twitter = read_file("/content/twitter.txt")
research_paper = read_file("/content/research_paper.txt")
```

Isolated spell correction: Jaccard Distance

```python
import nltk
import re
import string
from nltk.metrics.distance  import jaccard_distance
from nltk.util import ngrams
nltk.download('words')
from nltk.corpus import words
correct_words = words.words()

def correctSpellingInText(text):

  preprocessedWords = []

  for word in text:
    #remove HTML tags
    html_tags = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
    cleantext = re.sub(html_tags, '', word)

    #replace punctuation marks by space
    regular_punct = list(string.punctuation)
    for punc in regular_punct:
          if punc in cleantext:
              cleantext = cleantext.replace(punc, '')

    preprocessedWords.append(cleantext)

  
  print("preprocessedWords", preprocessedWords)
  print("cleantext",cleantext)

  spellCorrectedText = ""

  for word in preprocessedWords:
    if(len(word)>1 and not(word.isnumeric())):
      temp = [(jaccard_distance(set(ngrams(word, 2)),
                              set(ngrams(w, 2))),w)
            for w in correct_words if w[0]==word[0]]
      print(word + " -> " + sorted(temp, key = lambda val:val[0])[0][1])
      spellCorrectedText = spellCorrectedText + " " + sorted(temp, key = lambda val:val[0])[0][1]
    else:
      spellCorrectedText = spellCorrectedText + " " + word

  print(spellCorrectedText)

  return spellCorrectedText
```

Question 03: Lemmatization

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
nltk.download('punkt')
```

```python
wordLemmatizer = WordNetLemmatizer()

def lemmatizeText(text):
    tokenized_words = word_tokenize(text)
    lemmatized_word = [wordLemmatizer.lemmatize(word)
 for word in tokenized_words]
    return " ".join(lemmatized_word)

lemmatized_student_feedback = lemmatizeText(student_feedback)
lemmatized_twitter = lemmatizeText(twitter)
lemmatized_research_paper = lemmatizeText(research_paper)
```

```python
print("===========================Before Lemmatized Student Feedback=================")
print(student_feedback)
print("\n")
print ("==================== Lemmatized Student Feedback ===========================")
print(lemmatized_student_feedback)
```
